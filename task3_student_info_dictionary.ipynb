```python
#Implement a program to store student records (name, roll, marks,contact number) using a dictionary.
# a. Provide menu options to add, search, and display students

from IPython.display import clear_output
import os

print("Kripsan Thakuri's Work:\n ")

student_records = {}

def clear_screen():
    clear_output(wait=True)
    
def add_records():
    roll = int(input("Enter Student Roll no.: "))
    if roll in student_records:
        print("Roll no. already exists. Try another one.")
    else:
        print("Enter student info: \n ")
        name = input("Enter Student Name: ")
        marks = float(input("Enter Student Marks: "))
        contact = int(input("Enter Contact number: "))

        student_records[roll] = {
            "name": name,
            "marks": marks,
            "contact": contact
        }
        print("Data stored successfully!")

def search_records():
    roll = int(input("Enter the Roll no. of student: "))
    if roll not in student_records:
        print("Roll no. not found. Try again.")
            
    else:
        print(f"Student info for Roll no. {roll}: ")
        print(f"Name: {student_records[roll]['name']}" )
        print(f"Marks: {student_records[roll]['marks']}" )
        print(f"Contact: {student_records[roll]['contact']}" )

def display_records():
    if not student_records:
        print("No data found.")
            
    else:
        print("Student Details: \n")
        for roll, details in student_records.items():
            print(f"Roll no. : {roll}")
            print(f"Name: {details['name']}")
            print(f"Marks: {details['marks']}")
            print(f"Contact: {details['contact']} \n")
                
while True:
    #clear_screen()
    print("Welcome to student records!! \n")
    print("Your options are: ")
    print("1. Add Student info")
    print("2. Search student info")
    print("3. Display student info")
    print("4. Exit \n")
    choice = int(input("Enter a choice (1-4): "))

    if choice == 1:
        clear_screen()
        add_records()

    elif choice == 2:
        clear_screen()
        search_records()

    elif choice == 3:
        clear_screen()
        display_records()
                
    elif choice == 4:
        clear_screen()
        print('Exiting...')
        break

    else:
        print("Invalid input. Choose another option.")
        
        
        
                   





```
